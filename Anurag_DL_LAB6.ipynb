```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
words = [
    "Attention",
    "mechanism",
    "helps",
    "neural",
    "network",
    "models",
    "focus",
    "better"
]
words
```

```python
attention = np.array([
    [0.30, 0.10, 0.10, 0.10, 0.10, 0.10, 0.10, 0.10],
    [0.05, 0.40, 0.10, 0.10, 0.10, 0.10, 0.10, 0.05],
    [0.05, 0.10, 0.35, 0.15, 0.10, 0.10, 0.10, 0.05],
    [0.05, 0.10, 0.15, 0.35, 0.15, 0.10, 0.05, 0.05],
    [0.05, 0.10, 0.10, 0.15, 0.35, 0.15, 0.05, 0.05],
    [0.05, 0.10, 0.10, 0.10, 0.15, 0.35, 0.10, 0.05],
    [0.05, 0.10, 0.10, 0.05, 0.05, 0.15, 0.40, 0.10],
    [0.05, 0.05, 0.10, 0.05, 0.05, 0.10, 0.15, 0.45]
])
attention
```

```python
attention.sum(axis=1)
```

```python
plt.figure(figsize=(7, 5))
sns.heatmap(
    attention,
    xticklabels=words,
    yticklabels=words,
    annot=True,
    fmt='.2f',
    cmap = 'Blues'
)
plt.title('Attention Heatmap')
plt.xlabel('Key (Words being attended to)')
plt.ylabel('Query (Current word)')
plt.tight_layout()
plt.show()
```

```python
for i, query_word in enumerate(words):
    max_index = np.argmax(attention[i])
    print(f'Query word "{query_word}" attends most to "{words[max_index]}" with score {attention[i, max_index]:.2f}')
```
